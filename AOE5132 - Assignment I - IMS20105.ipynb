### QUESTION 1

### a) Make a pie chart of the masses of the various planets present in our solar system with a clearly visible legend & percentages. 


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('Planet_mass.csv')

# Creating plot
fig = plt.figure(figsize = (10,7))
pie = plt.pie(data['Mass_Mearth'], autopct='%1.2f%%', pctdistance=1.1, explode=[0.4, 1.4, 1.8, 2.2,0,0,0,0])
plt.legend(pie[0], labels  = data['Planet'], loc = 'best', bbox_to_anchor = (1,1))
plt.show()
```

### QUESTION 2

### (b) Make a bar chart of the masses of the various planets of our solar system showing the percentage of the mass distribution.


```python
fig = plt.figure(figsize = (10, 5))

# calculating Mass %
percent_mass = []
for i in range(data['Mass_Mearth'].shape[0]):
    p = data['Mass_Mearth'][i] / sum(data['Mass_Mearth']) * 100
    percent_mass.append(p)

# creating the bar plot
bars = plt.bar(data['Planet'], data['Mass_Mearth'], width = 0.4)
plt.yscale('log')
plt.xlabel("Planets in the Solar System",fontsize=20)
plt.ylabel("Mass (M_Earth)",fontsize=20)
plt.title("Mass distribution among planets in our solar system")
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)

# Adding percentage labels on top of the bars
for bar, percent in zip(bars, percent_mass):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percent:.2f}%', 
             ha='center', va='bottom', fontsize=12, color='black')

plt.tight_layout()  # Ensures that labels are not cut off
plt.show()
```

### (d) Plot the light curve of the gamma ray burst (GRB) using the .lc fits file shared with you. The trigger time of the GRB is 365734463.936 sec. 

```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

#Event file name  ############Detector B1###########
event_filename ='swift_bat_0.1s_15_150_120804A.lc'

#Open the fits file
pha_list = fits.open(event_filename, memmap=True)

#Get info of the fits file
pha_list.info()
#Print the columns of the header
print(pha_list[1].columns)
# load the data in separate variable
pha_data1 = pha_list[1].data

#List of Time
Time = pha_data1['TIME']
Trigger_time = 365734463.936
t_lc = np.subtract(Time, Trigger_time)
rate_lc = pha_data1['RATE']

#Light curve plot
fig_batlc, ax2 = plt.subplots(figsize=(8,7))
ax2.step(t_lc,rate_lc, 'k')
ax2.set_xlim([-4,9])
ax2.tick_params(axis='both',which='both', width=1,direction='in')
ax2.tick_params(which='major', length=8)
ax2.tick_params(which='minor', length=4)
ax2.tick_params(labelsize=15)
ax2.tick_params(axis='x', labelsize=18)
ax2.tick_params(axis='y', labelsize=18)
ax2.set_ylabel('Counts/bin/det', fontsize=21)
ax2.set_xlabel('Time since Swift BAT trigger(s)', fontsize=21)
plt.show()

```

### QUESTION 5

### (e) Make the plot of the time evolution of the sunspot using the data shared with you. Is there a periodicity observed in the data? If so, how will you estimate the periodicity?


```python
import pandas as pd
import matplotlib.pyplot as plt

sunspots = pd.read_csv('Sunspots.csv')

# Convert the 'Date' column to datetime format
sunspots['Date'] = pd.to_datetime(sunspots['Date'])

# Plot the time evolution
plt.figure(figsize=(10, 6))
plt.plot(sunspots['Date'], sunspots['Monthly_Mean_Total_Sunspot_Number'], marker='o', linestyle='-')
plt.title('Time Evolution trends of Sunspots',fontsize = 20)
plt.xlabel('Date',fontsize = 15)
plt.ylabel('Monthly Mean Total Sunspot Number',fontsize = 15)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from scipy.fft import fft, fftfreq
# Calculating the time interval
time_interval = ((sunspots["Date"].iloc[-1] - sunspots["Date"].iloc[0]).days)/365

# Creating the FFT
sunspot_series = sunspots["Monthly_Mean_Total_Sunspot_Number"].values
fft_values = np.fft.fft(sunspot_series)
fft_freqs = np.fft.fftfreq(len(sunspot_series), d=time_interval / len(sunspot_series))

# FFT power spectrum
plt.figure(figsize=(10, 6))
plt.plot(1 / fft_freqs, np.abs(fft_values))
plt.title('FFT Power Spectrum of Sunspot trends')
plt.xlabel('Period (years)')
plt.ylabel('Amplitude')

plt.xlim(0,300) 

plt.grid(True)
plt.show()
```

### FFT decomposes the time series into its constituent frequencies, revealing the dominant periodic components. In the context of sunspot analysis, these frequencies correspond to the different cycles and periodicities observed in solar activity.

```python
from astropy.table import Table
import matplotlib.pyplot as plt
t1 = Table.read('Fermi_GBM_catalog_15_8_22.txt',format='ascii',guess=False, delimiter='|')

T90=t1['t90']
fluence =t1['fluence']

# Creating frequency histogram
fig, ax = plt.subplots(figsize =(9, 7))
fluence_hist = ax.hist(fluence, bins = 20)
plt.title("Fluence of GRB")
plt.show()

fig, ax = plt.subplots(figsize =(9, 7))
T90_hist = ax.hist(T90, bins = 20)
plt.title("T90 of GRB")
plt.show()
```

```python
# Viewing the contents of the histogram
print("fluence_hist:",fluence_hist)

print("T90_hist:",T90_hist)
```

#### FLUENCE

```python
import pandas as pd

# Extract class intervals and frequencies
frequencies = fluence_hist[0]
bin_edges = fluence_hist[1]

# Create a DataFrame
table_fluence = {'Class Intervals': [], 'Frequencies': []}

for i in range(len(bin_edges) - 1):
    interval = f"{bin_edges[i]:.2f} - {bin_edges[i+1]:.2f}"
    table_fluence['Class Intervals'].append(interval)
    table_fluence['Frequencies'].append(frequencies[i])

df = pd.DataFrame(table_fluence)

# Print the DataFrame
print(df)

```
